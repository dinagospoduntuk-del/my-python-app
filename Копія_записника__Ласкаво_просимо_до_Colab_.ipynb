```
from transformers import pipeline

# Ініціалізація конвеєра аналізу настроїв
sentiment_analyzer = pipeline('sentiment-analysis')

# Приклад тексту для аналіз
text1 = "Я люблю цей фільм, він чудовий!"
text2 = "Цей продукт жахливий і мене дуже розчарував."
text3 = "Я не впевнений, що думаю про це."

# Виконання аналізу настроїв
result1 = sentiment_analyzer(text1)
result2 = sentiment_analyzer(text2)
result3 = sentiment_analyzer(text3)

# Виведення результатів
print(f"Текст 1: '{text1}'\nРезультат: {result1}\n")
print(f"Текст 2: '{text2}'\nРезультат: {result2}\n")
print(f"Текст 3: '{text3}'\nРезультат: {result3}\n")
```

```
# Завантажуємо модель генерації тексту
generator = pipeline('text-generation', model='gpt2')

# Задаємо початок фрази, а модель має її продовжити
prompt = "In the future, artificial intelligence will"

output = generator(prompt, max_length=50, num_return_sequences=1)

print(output[0]['generated_text'])
```
